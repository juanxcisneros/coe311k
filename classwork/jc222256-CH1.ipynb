```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace (0, np.pi/2,100)
y=np.sin(x)
plt.subplot(2, 2, 1)
plt.plot(x, np.arcsin(x), 'r')
plt.subplot(2,2,2)
plt.plot(y, "b--")
plt.show()
plt.subplot(2,2,3)
plt.scatter(x,y,2,'g')
plt.show()
z = np.random.rand(1000)
plt.subplot(2,2,4)
plt.hist(z, 25)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# function
def f(x):
    return np.sin(x)

# interval
a = 0
b = 2 * np.pi

# number of points
N = 50

# x values
x = np.linspace(a, b, N)

#step size
h = (b - a) / (N - 1)

#true derivative
true_derivative = np.cos(x)

#central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

#forward Difference Approximation
forward = (f(x + h) - f(x)) / h

#backward Difference Approximation
backward = (f(x) - f(x - h)) / h

#plot

plt.figure()

#true derivative
plt.subplot(2,2,1)
plt.plot(x, true_derivative)
plt.title("cos(x)")

#forward difference
plt.subplot(2,2,2)
plt.plot(x, forward)
plt.title("Forward difference")

#backward difference
plt.subplot(2,2,3)
plt.plot(x, backward)
plt.title("Backward difference")

#central difference
plt.subplot(2,2,4)
plt.plot(x, df_dx)
plt.title("Central difference")

plt.tight_layout()
plt.show()
```

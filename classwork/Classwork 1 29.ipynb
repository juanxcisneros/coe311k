```python
import random
n = 1_000_000
inside = 0
i = 0

while (i < n):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    if x**2 + y**2 <= 1:
        inside += 1
    i += 1

pi_est = 4 * inside / n
print(pi_est)
```

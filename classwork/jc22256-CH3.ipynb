```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monte_carlo_pi(n):
    inside = 0
    i = 0
    while i < n:
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
        i += 1
    return 4 * inside / n
```

```python
n_samples = 1000
n_trials = 100
pi_true = np.pi

abs_errors = []
rel_errors = []

for _ in range(n_trials):
    pi_est = monte_carlo_pi(n_samples)
    abs_err = abs(pi_true - pi_est)
    rel_err = abs((pi_true - pi_est) / pi_true)
    abs_errors.append(abs_err)
    rel_errors.append(rel_err)
```

```python
plt.figure()
plt.plot(abs_errors)
plt.xlabel("Trial number")
plt.ylabel("Absolute error")
plt.title("Absolute Error")
plt.grid(True)
plt.show()
```

```python
plt.figure()
plt.plot(rel_errors)
plt.xlabel("Trial number")
plt.ylabel("Relative error")
plt.title("Relative Error")
plt.grid(True)
plt.show()
```

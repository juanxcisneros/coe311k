```python
import numpy as np
import matplotlib.pyplot as plt

# function
def f(x, y):
    return -2*y

x_start = 0
x_end = 2

step_sizes = np.arange(0.01, 1.01, 0.05)

#first run
h = step_sizes[0]

num_steps = int((x_end - x_start) / h)

x_base = np.linspace(x_start, x_end, num_steps + 1)
y_base = np.zeros(num_steps + 1)

y_base[0] = 1

# euler method
for i in range(num_steps):
    y_base[i+1] = y_base[i] + h * f(x_base[i], y_base[i])

print("Comparing to first run (h = 0.01)\n")

# plot
plt.plot(x_base, y_base, label="h = 0.01")

#loop
for h in step_sizes[1:]:
    
    num_steps = int((x_end - x_start) / h)
    
    x_vals = np.linspace(x_start, x_end, num_steps + 1)
    y_vals = np.zeros(num_steps + 1)
    
    y_vals[0] = 1
    
    #euler method
    for i in range(num_steps):
        y_vals[i+1] = y_vals[i] + h * f(x_vals[i], y_vals[i])
    
    #compare
    y_compare = np.interp(x_vals, x_base, y_base)
    
    #calculate error
    errors = []
    total_error = 0
    
    for j in range(len(y_vals)):
        e = abs(y_vals[j] - y_compare[j])
        errors.append(e)
        total_error += e
    
    print("h =", round(h, 2))
    print("pointwise error:", errors)
    print("total error:", total_error)
    print()
    
    #plot
    plt.plot(x_vals, y_vals, label="h=" + str(round(h,2)))

# plot
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler method for dy/dx = -2y")
plt.show()
```

```python
# put your grades here (out of 100)
attendance = 100
quizzes = 96.67

# exams (just type your 4 grades here)
exams = [72.5, 75, 81.67, 82]

# sort exams highest to lowest
exams.sort(reverse=True)

# weights for exams (highest → lowest)
weights = [0.25, 0.23, 0.17, 0.15]

# calculate exam total
exam_total = 0
for i in range(4):
    exam_total += exams[i] * weights[i]

# final grade
final_grade = (attendance * 0.08) + (quizzes * 0.12) + exam_total

# attendance penalty
if attendance < 50:
    final_grade -= 3  # simple minus drop

print("Final numeric grade:", final_grade)

# letter grade
if final_grade >= 90:
    letter = "A"
elif final_grade >= 80:
    letter = "B"
elif final_grade >= 65:
    letter = "C"
elif final_grade >= 50:
    letter = "D"
else:
    letter = "F"

print("Letter grade:", letter)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Naive matrix multiplication from class
def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]

    return C

# Build matrix sizes: 3x3, 6x6, 12x12, ...
sizes = []
size = 3
for i in range(8):
    sizes.append(size)
    size = size * 2

naive_times = []
numpy_times = []
matrix_numbers = list(range(1, 9))

# Generate 8 pairs of random matrices and time both methods
for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    # Convert to lists for naive function
    A_list = A.tolist()
    B_list = B.tolist()

    # Time naive multiplication
    start = time.perf_counter()
    C_naive = naive_matrix_multiplication(A_list, B_list)
    end = time.perf_counter()
    naive_times.append(end - start)

    # Time NumPy multiplication
    start = time.perf_counter()
    C_numpy = np.matmul(A, B)
    end = time.perf_counter()
    numpy_times.append(end - start)

# Print runtimes
print("Matrix Number   Matrix Size   Naive Time (s)   NumPy Time (s)")
for i in range(8):
    print(f"{i+1:^13} {sizes[i]}x{sizes[i]:<8} {naive_times[i]:<16.6f} {numpy_times[i]:.6f}")

# Plot runtimes
plt.plot(matrix_numbers, naive_times, marker='o', label='Naive')
plt.plot(matrix_numbers, numpy_times, marker='o', label='NumPy')

plt.title("Naive vs NumPy Matrix Multiplication Runtime")
plt.xlabel("Matrix Number")
plt.ylabel("Runtime (seconds)")
plt.xticks(matrix_numbers)
plt.legend()
plt.grid(True)
plt.show()
```

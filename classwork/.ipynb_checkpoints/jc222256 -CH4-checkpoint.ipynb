```python
import numpy as np

pi_true = np.pi

sum_value = 0
n = 0

while n < 5000:
    
    if n % 2 == 0:
        sign = 1
    else:
        sign = -1
    
    denominator = 2 * n + 1
    term = sign / denominator
    sum_value = sum_value + term
    
    pi_estimate = 4 * sum_value
    
    if (n + 1) % 100 == 0:
        error = abs(pi_true - pi_estimate)
        print("After", n + 1, "terms:")
        print("Estimated pi =", pi_estimate)
        print("Error =", error)
        print("")
    
    n = n + 1
```

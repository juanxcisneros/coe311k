```python
import numpy as np
import matplotlib.pyplot as plt

# constants
k = 1.0
h_values = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]

x_start = 0
x_end = 10
y0 = 1.0
v0 = 0.0

plt.figure(figsize=(10, 6))

for h in h_values:
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y0
    v_values[0] = v0

    for i in range(1, len(x_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)

    plt.plot(x_values, y_values, label=f"h = {h}")

plt.title("Euler's Method")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#when does it become unstable?
#immediately for any positive h. Becomes visibly unstable once h is bigger

#is it becoming unstable?
#Yes

#Can Euler solve this with a larger h?
#No, larger h makes it worse
```

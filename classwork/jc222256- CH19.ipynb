```python
import numpy as np
import matplotlib.pyplot as plt

h = 0.0001

t0 = 0
tf = 1

t = np.arange(t0, tf + h, h)

y = np.zeros(len(t))

y[0] = 0

def f(t, y):
    return -1000*y + 3000 - 2000*np.exp(-t)

for i in range(len(t)-1):
    y[i+1] = y[i] + h * f(t[i], y[i])
# analytical solution
y_exact = 3 - 0.998*np.exp(-1000*t) - 2.002*np.exp(-t)

# plot
plt.plot(t, y, label="euler")
plt.plot(t, y_exact, '--', label="exact")

plt.xlabel("t")
plt.ylabel("y")
plt.title("euler vs analytical")
plt.legend()
plt.grid()

plt.show()
```

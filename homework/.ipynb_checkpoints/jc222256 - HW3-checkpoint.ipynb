```python
import numpy as np
import matplotlib.pyplot as plt

x_values = []

x = 0
while x <= 10:
    x_values.append(x)
    x = x + 0.1


exact_values = []
approx_1 = []
approx_2 = []
approx_3 = []
approx_4 = []

i = 0
while i < len(x_values):

    x = x_values[i]

    exact = np.sin(x)
    exact_values.append(exact)

    value1 = x
    approx_1.append(value1)

    value2 = x - (x**3)/6
    approx_2.append(value2)

    value3 = x - (x**3)/6 + (x**5)/120
    approx_3.append(value3)

    value4 = x - (x**3)/6 + (x**5)/120 - (x**7)/5040
    approx_4.append(value4)

    i = i + 1


error_1 = []
error_2 = []
error_3 = []
error_4 = []

i = 0
while i < len(x_values):

    error_1.append(abs(exact_values[i] - approx_1[i]))
    error_2.append(abs(exact_values[i] - approx_2[i]))
    error_3.append(abs(exact_values[i] - approx_3[i]))
    error_4.append(abs(exact_values[i] - approx_4[i]))

    i = i + 1


plt.plot(x_values, error_1)
plt.plot(x_values, error_2)
plt.plot(x_values, error_3)
plt.plot(x_values, error_4)

plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Error in Taylor Approximation of sin(x)")
plt.show()

#As more terms are included in the Taylor series, the truncation error generally decreases because the polynomial approximation becomes closer to the true sine function. 
#However, the error does not decrease uniformly for all values of x. 
#For small values of x (near 0), even one or two terms provide a very accurate approximation because higher-order terms are very small. 
#For larger values of x, the error is significantly larger when only a few terms are used, since higher powers of x grow quickly and the neglected terms become more important. 
#Therefore, the truncation error behaves much better for small x and requires more terms to maintain accuracy as x increases.
```
